```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
salario = np.array([100,200,300,500,400,150])
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c='black')
```

```python
plt.plot(salario, c='black',ls='--')
```

```python
plt.plot(salario, ls='--', marker='s')
```

```python
plt.plot(salario, c='black', ls='--', marker='s',ms=18)
plt.show()
```

```python
salarios_lucas = np.array([100,200,400,300])
salarios_gileno = np.array([50, 300, 500, 450])
```

```python
plt.plot(salarios_lucas, c='black',ls='--',marker='s',ms=8, label='Salários Lucas')
plt.plot(salarios_gileno, c='red',ls='--',marker='s',ms=8, label='Salários Gileno')
plt.legend()
plt.show()
```

```python
plt.plot(salarios_lucas, c='black',ls='--',marker='s',ms=8, label='Salários Lucas')
plt.plot(salarios_gileno, c='red',ls='--',marker='^',ms=8, label='Salários Gileno')
plt.legend()
plt.show()
```

```python
plt.plot(salarios_lucas, c='black',ls='--',marker='s',ms=8, label='Salários Lucas')
plt.plot(salarios_gileno, c='red',ls='--',marker='s',ms=8, label='Salários Gileno')
plt.legend(loc='upper right')
plt.show()
```

```python
#colocando mais uma linha
salarios_pedro = [200,150,300,700]
plt.plot(salarios_lucas, c='black',ls='--',marker='s',ms=10, label='Salários Lucas')
plt.plot(salarios_gileno, c='red',ls='--',marker='^',ms=10, label='Salários Gileno')
plt.plot(salarios_pedro, c='green',ls='--',marker='o',ms=10, label='Salários Pedro')
plt.legend()
plt.show()
```
