```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train_shoes.csv')
test = pd.read_csv('test_shoes.csv')
train.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
cols = ['shoe size','height']
cols2 = ['class']
x_train = train.as_matrix(cols)
y_train = train.as_matrix(cols2)
x_test = test.as_matrix(cols)
y_test = test.as_matrix(cols2)
y_train
```

```python
knn = KNeighborsClassifier(n_neighbors=3, weights='distance')
knn.fit(x_train, y_train.ravel())
```

```python
output = knn.predict(x_test)
```

```python
print(knn.predict([x_test[0]]))
print(y_test[0])
```

```python
correct = 0.0
for i in range(len(output)):
    if y_test[i][0] == output[i]:
        correct += 1
correct /len(output)
```
