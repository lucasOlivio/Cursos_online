```python
import numpy as np
```

```python
# x entradas e y saidas
x = np.genfromtxt('balance-scale.data',delimiter=',',usecols=(1,2,3,4))
y = np.genfromtxt('balance-scale.data',delimiter=',',usecols=(0))
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=42)
x_treino
```

```python
x_teste
```

```python
y_teste[0]
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=17)
knn.fit(x_treino, y_treino)
labels = knn.predict(x_teste)
labels
```

```python
y_teste
```

```python
np.sum(labels == y_teste)
```

```python
(labels == y_teste).sum()
```

```python
#porcentual de acerto
100 * (labels == y_teste).sum()/len(x_teste)
```

```python
knn.score(x_teste, y_teste)
```
