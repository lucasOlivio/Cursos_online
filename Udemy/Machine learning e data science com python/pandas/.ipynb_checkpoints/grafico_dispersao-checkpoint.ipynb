```python
iris = pydataset.data('iris')
```

```python
iris.head()
```

```python
plt.scatter(iris['Sepal.Length'], iris['Sepal.Width'], sizes=10*iris['Petal.Length'])
plt.show()
```

```python
def specie_color(x):
    if x=='versicolor':
        return 0
    elif x=='setosa':
        return 1
    return 2
```

```python
iris['SpeciesNumber'] = iris['Species'].apply(specie_color)
```

```python
plt.scatter(
    iris['Sepal.Length'], iris['Sepal.Width'], sizes = 15 * iris['Petal.Length'],
    c = iris['SpeciesNumber'], cmap='viridis', alpha=0.5
)
plt.show()
```
